# Exercises

## 1.

In the cell below, create a DataFrame `fruits` that looks like this:

| | Apples | Bananas |
|---|---|---|
| 0 | 30 | 21 |

````python
# Your code goes here. Create a dataframe matching the above diagram and assign it to the variable fruits.
fruits = pd.DataFrame({"Apples": [30], "Bananas": [21]})
fruits
````

**Key insight:** `pd.DataFrame(dict)` — keys become column names, values (lists) become column data. The row index defaults to 0, 1, 2, ... if `index=` is not specified.

## 2.

Create a dataframe `fruit_sales` that matches the diagram below:

| | Apples | Bananas |
|---|---|---|
| 2017 Sales | 35 | 21 |
| 2018 Sales | 41 | 34 |

````python
# Your code goes here. Create a dataframe matching the above diagram and assign it to the variable fruit_sales.
fruit_sales = pd.DataFrame({"Apples": [35, 41], "Bananas": [21, 34]}, index=["2017 Sales", "2018 Sales"])
fruit_sales
````

**Alternative solution:**
```python
fruit_sales = pd.DataFrame([[35, 21], [41, 34]], columns=['Apples', 'Bananas'],
                index=['2017 Sales', '2018 Sales'])
```

**Key insight:** A DataFrame can be built from a **dict of lists** (column-oriented) or a **list of lists** (row-oriented, needs `columns=` arg). The `index` parameter sets custom row labels.

## 3.

Create a variable `ingredients` with a Series that looks like:

```
Flour     4 cups
Milk       1 cup
Eggs     2 large
Spam       1 can
Name: Dinner, dtype: object
```

````python
ingredients = pd.Series(["4 cups", "1 cup", "2 large", "1 can"], index=["Flour", "Milk", "Eggs", "Spam"], name="Dinner")
ingredients
````

**Key insight:** `pd.Series(list, index=[...], name=...)` — `index` sets the row labels, `name` sets the Series name (shown on the last output line). The `name` attribute matters when a Series is used as a DataFrame column.

## 4.

Read a CSV dataset of wine reviews into a DataFrame called `reviews`. The dataset columns are:
`country`, `description`, `designation`, `points`, `price`, `province`, `region_1`, `region_2`, `variety`, `winery`

The CSV has an unnamed first column that should be used as the row index — use `index_col=0` when reading.

A sample CSV is provided below so you can practice `pd.read_csv()` with `index_col=0` without needing the full dataset file.

````python
import io

# Sample wine reviews data — same structure as the full winemag dataset
csv_data = (
    ",country,description,designation,points,price,province,region_1,region_2,variety,winery\n"
    "0,US,Tremendous varietal wine from Napa.,Martha's Vineyard,96,235.0,California,Napa Valley,Napa,Cabernet Sauvignon,Heitz\n"
    "1,Spain,Ripe aromas of fig and blackberry.,Carodorum Seleccion Reserva,96,110.0,Northern Spain,Toro,,Tinta de Toro,Bodega Carmen Rodriguez\n"
    "2,France,Salmon shade with scents of peaches.,Grand Brut Rose,90,52.0,Champagne,Champagne,,Champagne Blend,Gosset\n"
    "3,Italy,More Pinot Grigios should taste like this.,,90,15.0,Northeastern Italy,Alto Adige,,Pinot Grigio,Alois Lageder"
)

reviews = pd.read_csv(io.StringIO(csv_data), index_col=0)
reviews
````

**Key insight:** `pd.read_csv(filepath, index_col=0)` — `index_col=0` tells pandas to use the **first column** as the row index instead of creating a default 0, 1, 2, ... integer index.

The original winemag CSV has an unnamed leading column of row numbers — `index_col=0` absorbs it as the DataFrame index, matching the intended data shape.

To read from a local file, replace `io.StringIO(csv_data)` with the file path string:
```python
reviews = pd.read_csv("path/to/winemag-data_first150k.csv", index_col=0)
```
