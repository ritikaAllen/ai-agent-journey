```python
# 1D vs 2D vs 3D
arr_1d = np.array([1, 2, 3, 4, 6])
arr_2d = np.array([[1, 2, 3],
                   [4, 5, 6]])
arr_3d = np.array([[[1, 2], [3, 4]],
                   [[5, 6], [7, 8]]])

print("1D shape:", arr_1d.shape)
print("2D shape:", arr_2d.shape)
print("3D shape:", arr_3d.shape)

print("\n1D ndim:", arr_1d.ndim)
print("2D ndim:", arr_2d.ndim)
print("3D ndim:", arr_3d.ndim)

print("\nTotal elements in 2D:", arr_2d.size)
print("Data type:", arr_2d.dtype)
```

```python
arr = np.arange(12)
print("Original:", arr)
print("Shape:", arr.shape)

# Reshape into different configurations
as_2d   = arr.reshape(3, 4)
as_2d_b = arr.reshape(4, 3)
as_3d   = arr.reshape(2, 2, 3)

print("\nAs (3,4):\n", as_2d)
print("\nAs (4,3):\n", as_2d_b)
print("\nAs (2,2,3):\n", as_3d)
```

```python
arr = np.arange(12)

r1 = arr.reshape(3, -1)   # 3 rows, NumPy picks columns
r2 = arr.reshape(-1, 4)   # NumPy picks rows, 4 columns
r3 = arr.reshape(2, -1, 3)  # NumPy picks the middle dimension

print("reshape(3,-1):", r1.shape)
print("reshape(-1,4):", r2.shape)
print("reshape(2,-1,3):", r3.shape)
```
