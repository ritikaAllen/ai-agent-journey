## Exercise 2 — Create a null vector of size 10

**Concept:** `np.zeros` creates a float64 array by default — notice the `0.` output, not `0`.

````python
a = np.zeros(10)
print(a)
````

**Key insight:** `np.zeros` produces `float64` by default — values display as `0.` not `0`.  
To get integers: `np.zeros(10, dtype=int)`

---
## Exercise 3 — Create a null vector of size 10, set the 5th element to 1

**Concept:** Zero-based indexing — the 5th element is at index 4.

````python
a[4] = 1
print(a)
````

**Key insight:** Zero-based indexing — "5th element" = index 4. The value `1` is stored as `1.` because the array dtype is float64.

## Exercise 4 — Create a vector with values ranging from 10 to 49

**Concept:** `np.arange(start, stop)` — stop is **exclusive**.

````python
a = np.arange(10, 50)
print(a)
print(a.dtype)
````

**Key insight:** `np.arange(start, stop, step=1)` — the stop value is **exclusive**.  
Returns `int64` by default (no decimal points).

---
## Exercise 5 — Reverse a vector (first element becomes last)

**Concept:** Slicing with `[::-1]` reverses any array.

````python
a = np.arange(10, 50)
print(a[::-1])
````

**Key insight:** `[::-1]` means step = -1, so NumPy walks the array backwards.  
Same pattern works on Python lists too.

## Exercise 6 — Create a 3x3 matrix with values 0 to 8

**Concept:** `np.arange` + `reshape` to build a 2D matrix.

````python
a = np.arange(9).reshape(3, 3)
print(a)
````

**Key insight:** `np.arange(9)` produces `[0..8]`, and `reshape(3,3)` arranges them row by row.  
The grid looks like:
```
0  1  2
3  4  5
6  7  8
```

## Exercise 7 — Find indices of non-zero elements

**Concept:** `np.nonzero` returns the **indices** of non-zero elements (not the values themselves).

````python
a = np.array([1, 2, 0, 0, 4, 0])
print(np.nonzero(a))
````

**Key insight:** `np.nonzero` returns a **tuple** of arrays — one per dimension.  
For a 1D array: `(array([0, 1, 4]),)` → non-zero values live at indices 0, 1, and 4.  
The tuple structure supports multi-dimensional arrays (one index array per axis).

Difference from boolean masking:
- Boolean mask: `a[a != 0]` → returns the **values** `[1, 2, 4]`
- `np.nonzero(a)` → returns the **indices** `(array([0, 1, 4]),)`

## Exercise 8 — Create a 3x3 identity matrix

**Concept:** `np.eye(n)` creates an identity matrix — 1s on the diagonal, 0s elsewhere.

````python
a = np.eye(3)
print(a)
````

**Key insight:** The identity matrix is fundamental in linear algebra — multiplying any matrix by it returns the original matrix (like multiplying a number by 1).  
`np.eye` produces `float64` by default.

Grid:
```
1.  0.  0.
0.  1.  0.
0.  0.  1.
```

## Exercise 9 — Create a 3x3 array with random values, print min, max, mean

**Concept:** `np.random.random` generates floats in `[0, 1)`. Aggregate operations on random data.

````python
a = np.random.random((3, 3))
print(a)
print(a.min(), a.max(), a.mean())
````

**Key insight:** `np.random.random((3,3))` — the argument is a **shape tuple**, not two separate ints.  
All values fall in `[0.0, 1.0)`, so:
- `min` ≈ close to 0
- `max` ≈ close to 1
- `mean` ≈ ~0.5 (averages out over many values)

---
## Exercise 10 — Create a random array and subtract the mean (mean centering)

**Concept:** Mean centering — shifting data so the mean becomes 0.

````python
a = np.random.random((3, 3))
print(a - a.mean())
````

**Key insight:** Subtracting the mean from every element **centers** the data around 0.  
- Values that were below the mean → negative
- Values that were above the mean → positive
- The new mean ≈ 0 (tiny floating point error)

This is a foundational preprocessing step in ML — used in normalization, PCA, and neural network input scaling.
