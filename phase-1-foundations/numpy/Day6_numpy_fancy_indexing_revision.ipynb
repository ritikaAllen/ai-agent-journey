## Part 1: Fancy indexing basics

Indexing has two independent parts separated by the comma: `a[rows, cols]`.

- A **slice** (`1:`, `:2`) selects a contiguous range, in order.
- A **list** (`[0, 2]`) is *fancy indexing* — it selects exact positions, **in the order given**, and can repeat entries.

### Exercise 1: `a[1:, [0, 2]]`
Rows 1: → rows 1 and 2. Columns `[0, 2]` → column 0 then column 2, from those rows.


```python
a[1:, [0, 2]]
```

> **Key insight:** Fancy indexing with a list picks *exact positions in the order listed* — not a range, and not necessarily sorted. `[2, 0]` would reverse the column order.

### Exercise 2: `a[:2, [3, 1]]`
Two mistakes to watch for here:
1. `:2` is exclusive — rows 0 and 1 only, **not** row 2.
2. `[3, 1]` returns column 3 first, then column 1 — order matches the list, not ascending.


```python
a[:2, [3, 1]]
```

> **Key insight:** The exclusive-stop rule from `np.arange` (stop is not included) applies to *any* slice, including row/column slices in indexing.

### Exercise 3: `a[:1, [2, 0, 2]]`
Fancy indexing can repeat entries — `2` appears twice in the list, so column 2 appears twice in the output.


```python
a[:1, [2, 0, 2]]
```

> **Key insight:** Row 0 happens to equal its own column indices (`[0,1,2,3]`), so picking columns `2,0,2` gives values `2,0,2` — a coincidence of this particular array, not a general rule.

## Part 2: Integer index drops a dimension, slice/list keeps it

**Core rule:** An **integer** index says "give me exactly this one thing, drop the wrapper" → removes that axis.
A **slice or list** says "give me a sub-collection along this axis" → keeps the axis, even if only one element survives.

| Expression | Index type | Result shape | Axis dropped? |
|---|---|---|---|
| `a[0]` | integer | `(4,)` | row axis dropped |
| `a[0:1]` | slice | `(1, 4)` | row axis kept (size 1) |
| `a[:, 0]` | integer | `(3,)` | column axis dropped |
| `a[:, 0:1]` | slice | `(3, 1)` | column axis kept (size 1) |
| `a[0, 0]` | integer, integer | `()` | both dropped (scalar) |
| `a[[0]]` | list (fancy) | `(1, 4)` | row axis kept (size 1) — list behaves like slice |


```python
checks = {
    "a[0]": a[0],
    "a[0:1]": a[0:1],
    "a[:, 0]": a[:, 0],
    "a[:, 0:1]": a[:, 0:1],
    "a[0, 0]": a[0, 0],
    "a[[0]]": a[[0]],
}

for expr, result in checks.items():
    shape = np.shape(result)
    print(f"{expr:12s} -> shape {shape}")
```

## Summary

- **Fancy indexing** (passing a list) selects exact positions in the given order and allows repeats — unlike a slice, which is a contiguous, ordered range.
- **Exclusive stop** applies to every slice, not just `np.arange`.
- **Integer index → drops a dimension. Slice or list index → keeps it** (size 1 if only one element survives). This matters later for broadcasting and stacking, where consistent shapes matter.
