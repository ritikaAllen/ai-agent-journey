## Part 2 — Pandas

Run the setup cell below before attempting Pandas questions.

````python
import pandas as pd
import numpy as np

np.random.seed(42)
n = 120

country_pool = ['Italy', 'Portugal', 'US', 'Spain', 'France', 'Australia', 'Germany']
variety_pool = ['Pinot Noir', 'Chardonnay', 'Cabernet Sauvignon', 'Riesling',
                'Sangiovese', 'Shiraz', 'Tempranillo', 'White Blend']
region_pool = ['Napa Valley', 'Tuscany', 'Douro', 'Bordeaux', 'Barossa Valley', 'Rioja']
taster_names = ['Roger Voss', 'Paul Gregutt', "Kerin O'Keefe",
                'Anna Lee C. Iijima', 'Michael Schachner', 'Joe Czerwinski']
taster_handles = ['@vossroger', '@paulgwine', '@kerinokeefe',
                  '@annaiijima', '@wineschach', '@JoeCz']

countries = list(np.random.choice(country_pool, n))
varieties = list(np.random.choice(variety_pool, n))
points = list(np.random.randint(82, 99, n))
raw_prices = np.random.uniform(8, 250, n)
prices = [None if i % 6 == 0 else round(float(raw_prices[i]), 1) for i in range(n)]
regions = [None if i % 4 == 0 else np.random.choice(region_pool) for i in range(n)]

reviews = pd.DataFrame({
    'country': countries,
    'description': [f'Tasting notes for wine {i}.' for i in range(n)],
    'points': points,
    'price': prices,
    'region_1': regions,
    'region_2': [None if i % 5 == 0 else f'Sub-region {i % 6}' for i in range(n)],
    'taster_name': [taster_names[i % 6] for i in range(n)],
    'taster_twitter_handle': [taster_handles[i % 6] for i in range(n)],
    'title': [f'Winery {i % 15} {2010 + i % 10} {np.random.choice(variety_pool)} ' for i in range(n)],
    'variety': varieties,
    'winery': [f'Winery {i % 15}' for i in range(n)],
})

pd.set_option('display.max_rows', 5)
print(f'{len(reviews)} reviews · {reviews.shape[1]} columns')
print(f'Missing prices: {reviews.price.isnull().sum()}  |  Missing region_1: {reviews.region_1.isnull().sum()}')
reviews.head()
````
